```python
import numpy as np
import pandas as pd
```

```python
colleges = pd.read_csv("my_colleges_updt_1.csv")
colleges.head(20)
```

```python
fighters = pd.read_csv("my_fighters.csv")
fighters.head(20)
```

```python
# Group the DataFrame by "Primary Gym" and aggregate the fighter names into a list
grouped_gyms = fighters.groupby('Primary Gym')['Fighter'].agg(list).reset_index()
# Create a new DataFrame
gyms_fighters_df = pd.DataFrame(grouped_gyms)
gyms_fighters_df_clean = gyms_fighters_df.drop([18,20,21])
gyms_fighters_df_clean
```

```python
# Correct Marquez MMA
gyms_fighters_df_clean["Fighter"][16] = ["SEAN BRADY", "JOE PYFER", "PAT SABATINI", "ANDRE PETROSKI"] 
gyms_fighters_df_clean
```

```python
gym_lat_long = fighters.iloc[:,[4,9,10]]
gym_lat_long = gym_lat_long.drop_duplicates()
gym_lat_long
```

```python
# Perform left join based on "Primary Gym"
result_df = pd.merge(gyms_fighters_df_clean, gym_lat_long, on='Primary Gym', how='left')
result_df
```

```python
# Get rid of list in Fighter Column
result_df['Fighter'] = result_df['Fighter'].apply(lambda x: ', '.join(x))
result_df
```

```python
# Save the dataframe
result_df.to_csv("GYM LOCATIONS.csv", index=False)
```

```python
import folium
import pandas as pd

# Create a Folium map centered at the average coordinates
center_lat = result_df['Latitude'].mean()
center_lon = result_df['Longitude'].mean()
my_map = folium.Map(location=[center_lat, center_lon], zoom_start=7)

# Add markers for each college
for index, row in colleges.iterrows():
    # Format the popup with line breaks and bold headers using HTML
    popup_content = (
        f"<b>{row['School Name']}<br>"
    )
    
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_content,
        max_width=1200,  # Adjust the width as needed,
        icon=folium.Icon(color='blue')
    ).add_to(my_map)

# Add markers for each college
for index, row in result_df.iterrows():
    # Format the popup with line breaks and bold headers using HTML
    popup_content = (
        f"<b>Gym:</b> {row['Primary Gym']}<br>"
        f"<b>Fighters:</b> {row['Fighter']}"
    )
    
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=folium.Popup(popup_content, max_width=1200),
        icon=folium.Icon(color='red')
    ).add_to(my_map)



# Save the map as an HTML file
my_map.save('mma_map.html')
    
# Display the map
my_map
```
