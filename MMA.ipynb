```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def get_link_from_tapology(name):
    # Create a new instance of the Chrome driver
    driver = webdriver.Chrome()

    try:
        # Open the Tapology website
        driver.get("https://www.tapology.com/")

        # Find the search input field and enter the name
        search_input = driver.find_element(By.XPATH, '//*[@id="siteSearch"]')
        search_input.send_keys(name)

        # Find the search button and click it
        search_button = driver.find_element(By.XPATH, '//*[@id="search"]')
        search_button.click()

        # Wait for 3 seconds
        time.sleep(3)

        # Extract the link at the specified xpath
        link_xpath = '//*[@id="content"]/div[2]/table/tbody/tr[2]/td[1]/a'
        link_element = driver.find_element(By.XPATH, link_xpath)
        link = link_element.get_attribute("href")

        return link

    finally:
        # Close the browser window
        driver.quit()
```

```python
import re

def strip_substrings_with_apostrophes(data):
    # Define a regular expression pattern for substrings with apostrophes
    pattern = r'".*?"'

    # Use re.sub() to replace matched substrings with an empty string for each element in the list
    result = [re.sub(pattern, '', item) for item in data]

    return result


# Apply the function to the list
result = strip_substrings_with_apostrophes(names_result)

def trim_whitespace(items):
    # Use list comprehension to strip leading and trailing whitespace from each item
    trimmed_items = [item.strip() for item in items]
    return trimmed_items

def extract_substring_before_linebreak(strings):
    # Use list comprehension to extract substring before the first line break in each string
    substrings = [string.split("\n", 1)[0] for string in strings]
    return substrings

result1 =  trim_whitespace(result)
result2 =  extract_substring_before_linebreak(result1)
result3 = list(set(result2))
result3
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def get_all_urls(base_url, target_prefix, max_depth=3, current_depth=0):
    if current_depth > max_depth:
        return []

    all_urls = set()

    # Send a GET request to the base URL
    response = requests.get(base_url)
    if response.status_code != 200:
        print(f"Failed to fetch {base_url}. Status code: {response.status_code}")
        return all_urls

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all links on the page
    links = soup.find_all('a', href=True)

    # Extract URLs that start with the specified prefix and join them with the base URL
    for link in links:
        child_url = urljoin(base_url, link['href'])
        if child_url.startswith(target_prefix):
            all_urls.add(child_url)

    # Recursively crawl child URLs
    for child_url in all_urls.copy():
        all_urls.update(get_all_urls(child_url, target_prefix, max_depth, current_depth + 1))

    return all_urls

# Example usage:
root_url = "https://www.tapology.com/"
target_prefix = "https://www.tapology.com/fightcenter/fighters/"
all_urls_result = get_all_urls(root_url, target_prefix)

# Print the result
print("All URLs:")
for url in all_urls_result:
    print(url)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def get_links_from_xpath_pattern(url, xpath_pattern):
    # Create a new instance of the Chrome driver
    driver = webdriver.Chrome()

    try:
        # Open the webpage
        driver.get(url)

        # Wait for 2 seconds
        time.sleep(2)

        # Find all elements that match the XPath pattern
        elements = driver.find_elements(By.XPATH, xpath_pattern)

        # Extract href links from each element
        links = [element.get_attribute("href") for element in elements]

        return links

    finally:
        # Close the browser window
        driver.quit()

# Example usage:
url_input = "https://www.tapology.com/search?term=CHRIS+GUTIERREZ&search=Submit&mainSearchFilter=fighters"
xpath_pattern_input = '//*[@id="content"]/div[2]/table/tbody/tr[{}]/td[1]/a'

# Get links using the function
links_result = []
for i in range(2, 6):  # Assuming tr indices from 2 to 4
    try: 
        xpath_pattern = xpath_pattern_input.format(i)
        links_result.extend(get_links_from_xpath_pattern(url_input, xpath_pattern))
```

```python
import re

def split_string_on_numerical_value(file_path):
    try:
        # Read the contents of the file
        with open(file_path, 'r') as file:
            content = file.read()

        # Split the string using a regular expression
        split_list = re.split(r'\b\d{1,3}\b', content)

        # Remove empty strings from the list
        split_list = [item.strip() for item in split_list if item.strip()]

        return split_list
    except FileNotFoundError:
        print(f"Error: File not found at path {file_path}")
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None

# Example usage:
file_path = r"C:\Users\Bobby\Documents\Data\MMA\fighter_list.txt"
result_list = split_string_on_numerical_value(file_path)

result_list
```

```python
import pandas as pd

def read_colleges_data(file_path):
    """
    Read data from the colleges.txt file and create a dataframe.
    
    Parameters:
    - file_path: Path to the colleges.txt file.
    
    Returns:
    - Pandas DataFrame with columns ["School Name", "URL", "State"].
    """
    # Initialize lists to store data
    school_names = []
    urls = []
    states = []

    # Read data from the file
    with open(file_path, 'r') as file:
        for line in file:
            # Split each line into School Name, URL, and State
            data = line.strip().split('\t')
            if len(data) == 3:
                school_names.append(data[0])
                urls.append(data[1])
                states.append(data[2])

    # Create a dataframe
    df = pd.DataFrame({
        "School Name": school_names,
        "URL": urls,
        "State": states
    })

    return df

# Replace 'colleges.txt' with the actual path to your file
file_path = r"C:\Users\Bobby\Documents\Data\MMA\colleges.txt"

# Read data and create a dataframe
df = read_colleges_data(file_path)
df = df.iloc[1:,:]
df
```
